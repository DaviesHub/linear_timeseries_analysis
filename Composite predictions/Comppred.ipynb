```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

sns.set_style("darkgrid")
sns.mpl.rc("figure", figsize=(16, 6))
sns.mpl.rc("font", size=14)
```

```python
dig15pred = np.load("Dig15pred.npy")
dig16pred = np.load("Dig16pred.npy")
dig17pred = np.load("Dig17pred.npy")
dig18pred = np.load("Dig18pred.npy")
dig19pred = np.load("Dig19pred.npy")
dig20pred = np.load("Dig20pred.npy")
dig21pred = np.load("Dig21pred.npy")
dig22pred = np.load("Dig22pred.npy")
dig23pred = np.load("Dig23pred.npy")
dig24pred = np.load("Dig24pred.npy")
dig25pred = np.load("Dig25pred.npy")
dig15forecast = np.load("Dig15fcast.npy")
dig16forecast = np.load("Dig16fcast.npy")
dig17forecast = np.load("Dig17fcast.npy")
dig18forecast = np.load("Dig18fcast.npy")
dig19forecast = np.load("Dig19fcast.npy")
dig20forecast = np.load("Dig20fcast.npy")
dig21forecast = np.load("Dig21fcast.npy")
dig22forecast = np.load("Dig22fcast.npy")
dig23forecast = np.load("Dig23fcast.npy")
dig24forecast = np.load("Dig24fcast.npy")
dig25forecast = np.load("Dig25fcast.npy")
totsteamforecast = np.load("TotSteamfcast.npy")
valindex = np.load("DigHXindex.npy")
VTindex = np.load("ValTestIndex.npy")
Totalsteam = np.load("TotSteam.npy")
Totalsteampred = np.load("TotSteampred.npy")
```

```python
comp_preds = dig15pred + dig16pred + dig17pred + dig18pred + dig19pred + dig20pred + dig21pred + dig22pred + dig23pred + dig24pred + dig25pred
comp_fcast = dig15forecast + dig16forecast + dig17forecast + dig18forecast + dig19forecast + dig20forecast + dig21forecast + dig22forecast + dig23forecast + dig24forecast + dig25forecast
```

```python
plt.plot(valindex, comp_preds, color = "red")
plt.plot(valindex, Totalsteam, color = "blue")
plt.plot(valindex, Totalsteampred, color = "green")
plt.legend(['Composite Prediction', 'Actual', 'Total Steam Prediction'])
plt.title("Predictions vs Actual", size = 24)
plt.show()
```

```python
print("Composite MAE: %.4f" %mean_absolute_error(Totalsteam, comp_preds))
print("Composite RMSE: %.4f" %np.sqrt(mean_squared_error(Totalsteam, comp_preds)))
```

```python
print("Total Steam MAE: %.4f" %mean_absolute_error(Totalsteam, Totalsteampred))
print("Total Steam RMSE: %.4f" %np.sqrt(mean_squared_error(Totalsteam, Totalsteampred)))
```

```python
plt.plot(VTindex[220:], comp_fcast, color = "red")
plt.plot(valindex, Totalsteam, color = "blue")
plt.plot(VTindex[220:], totsteamforecast, color = "green")
plt.legend(['Composite Forecast', 'Actual', 'Total Steam Forecast'])
plt.title("Arima Forecasts", size = 24)
plt.show()
```
